```python
# Task 7 
import seaborn as sns

df_tips = sns.load_dataset("tips")
Q1 = df_tips['total_bill'].quantile(0.25)
Q3 = df_tips['total_bill'].quantile(0.75)

IQR = Q3 - Q1
Lower=Q1- (1.5*IQR)
Upper=Q3 + (1.5 * IQR)


df_tips.boxplot(column=["total_bill"])
plt.show()

df_tips.boxplot(column=["tip"])
plt.show()

outliers=df_tips[(df_tips['total_bill'] < Lower) | (df_tips['total_bill'] > Upper)]
print("Total Bill Outliers is ", outliers)

Q1_2 = df_tips['tip'].quantile(0.25)
Q3_2= df_tips['tip'].quantile(0.75)

IQR_2 = Q3_2 - Q1_2
Lower_2 = Q1_2 - (1.5*IQR_2)
Upper_2 = Q3_2 + (1.5 * IQR_2)

outliers_2 = df_tips[(df_tips['tip']<Lower_2) | (df_tips['tip'] > Upper_2)]
print("Total Bill Outliers is ", outliers_2)

```

```python
# Task 8 
import seaborn as sns

df_tips = sns.load_dataset("tips")

Q1 = df_tips['total_bill'].quantile(0.25)
Q3 = df_tips['total_bill'].quantile(0.75)
IQR = Q3 - Q1
Lower=Q1- (1.5*IQR)
Upper=Q3 + (1.5 * IQR)


Q1_2 = df_tips['tip'].quantile(0.25)
Q3_2= df_tips['tip'].quantile(0.75)
IQR_2 = Q3_2 - Q1_2
Lower_2 = Q1_2 - (1.5*IQR_2)
Upper_2 = Q3_2 + (1.5 * IQR_2)

print('Total Data ' , df_tips.shape)

outliers=df_tips[((df_tips['total_bill'] < Lower) | (df_tips['total_bill'] > Upper)) & ((df_tips['tip']<Lower_2) | (df_tips['tip'] > Upper_2))]

print('Rows Deleted ', outliers.shape)

```
